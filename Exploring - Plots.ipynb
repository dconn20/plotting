# Exploring Plots
***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("ggplot")
```

```python
plt.plot ([1,2,3], 'r-', label = 'red line')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot([1,2,3], 'r-', label = 'red line')
ax.legend()
```

#### from the seaborn tutorial
*****

```python
x = np.linspace(0.0, 100.0, 100)
y = x + np.random.normal(0.0, 10, len(x))
```

```python
sns.relplot(x=x, y=y);
```

<br>

#### Distplot
***

```python
x = np.random.normal(1, 2, 1000)
```

```python
sns.displot(x, kde=True);
```

#### Bar Chart
***

```python
makes = ['Toyota', 'Ford', 'Hyundai', 'Volkswagen', 'BMW', 'Audi']
cars = np.random.choice(makes, 6000)
list(cars)
```

```python
sns.countplot(x=cars);
```
